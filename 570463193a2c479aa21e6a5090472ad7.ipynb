## Welcome to your notebook.


#### Run this cell to connect to your GIS and get started:

```python
from arcgis.gis import GIS
gis = GIS("home")
```

### My First Map

**My Map of Los Angeles**

```python
my_first_map = gis.map("Los Angeles, CA")
```

```python
my_first_map
```

### Adding First Layer

**Adding a second layer showing U.S demographics median household income in the U.S**

```python
# Item Added From Toolbar
# Title: Esri Updated Demographics Variables 2023 | Type: Feature Service | Owner: esri_demographics
item = gis.content.get("ddd1510ed1964e2e8fb1ded93e9b03b1")
item
```

```python
my_first_map.content.add(item)
```

### Adding Second Layer 

**Adding second layer showing major cities in the U.S**

```python
# Item Added From Toolbar
# Title: USA Major Cities | Type: Feature Service | Owner: esri_dm
item2 = gis.content.get("9df5e769bfe8412b8de36a2e618c7672")
item2
```

```python
my_first_map.content.add(item2)
```

### Adding a Third Layer

```python
# Item Added From Toolbar
# Title: ACS Median Household Income Variables - Centroids | Type: Feature Service | Owner: esri_demographics
item3 = gis.content.get("cab3fe0ee8304888a47a58355a472904")
item3
```

```python
my_first_map.content.add(item3)
```

### Change Symbology of Second Layer

```python
rend_manager = my_first_map.content.renderer(1) #Get the Layer by index position 0 is the second layer
# Get the layer's renderer dataclass
simple_rend = rend_manager.renderer
```

```python
# simple_rend is a dataclass so you can make any changes to it and then set it on the renderer property
simple_rend.symbol.size = 5
rend_manager.renderer = simple_rend
```

```python
simple_rend.symbol.size = 10
rend_manager.renderer = simple_rend
```
